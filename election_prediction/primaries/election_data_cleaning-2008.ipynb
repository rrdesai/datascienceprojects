```python
import json 
import pandas as pd
import numpy as np
```

```python
f =  open('primaries2008_2.json').read()

try:
    j = json.loads(f)[-1]
except:
    j = json.loads(f)



```

```python
df = pd.DataFrame()
for i in j:
    temp = pd.read_json(json.dumps(j[i]), orient = 'records')
    temp = temp.transpose()
    temp['state'] = i
    df = pd.concat((df,temp))
```

```python
test = df.fillna('00')
```

```python
for i in test.columns:
    try:
        test[i] = test[i].apply(lambda x: int(x[0].replace(',','')))
    except:
        continue
```

```python
test = test.drop(['Dodd','Gravel','Hunter','Kucinich','Richardson','Thompson','Biden'], axis=1)
```

```python
test['total_votes'] = test.sum(axis=1)
test['dem'] = test.Clinton + test.Obama + test.Edwards
test['rep'] = test.Giuliani + test.Huckabee + test.McCain + test.Paul + test.Romney
```

```python
test['rep'] = test['rep']/test['total_votes']
test['dem'] = test['dem']/test['total_votes']
```
