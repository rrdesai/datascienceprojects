```python
import json 
import pandas as pd
import numpy as np
```

```python
f =  open('louisiana2016.json').read()

try:
    j = json.loads(f)[-1]
except:
    j = json.loads(f)



```

```python
df = pd.DataFrame()
for i in j:
    temp = pd.read_json(json.dumps(j[i]), orient = 'records')
    temp = temp.transpose()
    temp['state'] = i
    df = pd.concat((df,temp))
```

```python
test = df.fillna('00')
```

```python
for i in test.columns:
    try:
        test[i] = test[i].apply(lambda x: int(x[1].replace(',','')))
    except:
        continue
```
